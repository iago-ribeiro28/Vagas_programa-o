```python
import logging
import pandas as pd
from bs4 import BeautifulSoup
import requests
import os
from datetime import datetime
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36 Edg/137.0.0.0'}
url = f'https://www.linkedin.com/jobs-guest/jobs/api/seeMoreJobPostings/search?geoId=106057199&keywords=""&f_WT=1&position=3&start=0'
response = requests.get(url, headers= headers)
page_html = response.text
soup = BeautifulSoup(page_html, features='html.parser')
soup
```

```python
with open('soup_content.txt', 'w', encoding='utf-8') as f:
    f.write(str(soup))
```
