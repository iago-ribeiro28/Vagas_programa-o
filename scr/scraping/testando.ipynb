```python
import logging
import pandas as pd
from bs4 import BeautifulSoup
import requests
import os
from datetime import datetime
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36 Edg/137.0.0.0'}
```

```python
vagas = {
            'id': [],
            'modelo':[],
            'titulo':[],
            'tempo': [],
            'inscritos': []
        }

modelos = {1: 'Presencial',
           2: 'Remoto', 
           3: 'Hibrido'}
```

```python
# Criar uma lista para armazenar todos os job_ids
vagas = {
            'id': [],
            'modelo':[],
            'titulo':[],
            'tempo': [],
            'inscritos': []
        }
paginas = 0

for modelo in modelos: 
    print(f'O modelo atual é {modelo}') 
    while True:
        url = f'https://www.linkedin.com/jobs-guest/jobs/api/seeMoreJobPostings/search?geoId=106057199&keywords="Analista%20de%20dados"&f_WT={modelo}&position=3&start={paginas}'
        response = requests.get(url, headers= headers)
        page_html = response.text
        soup = BeautifulSoup(page_html, features='html.parser')
        divs = soup.find_all('div', class_='base-card')
        travou = soup.find_all('script', type='text/javascript')
        
        if len(divs) == 0 and len(travou) == 0:
            break
        
        elif len(travou) == 0:
            for card in divs:
                entity_urn = card.get('data-entity-urn')
                
                if entity_urn:  # Verificar se existe o atributo
                    id = entity_urn.split(':')[-1]
                    titulo = card.find('h3', class_='base-search-card__title').text.strip()
                    vagas['id'].append(id)
                    vagas['titulo'].append(titulo)
                    vagas['modelo'].append(modelos.get(modelo))
            paginas = paginas + 1
    print(f"O número de paginas percorridas para o modelo {modelo} foram {paginas}")

# Exibir todos os IDs
print(vagas['id'])
print(vagas['modelo'])
print(vagas['titulo'])
print(f"O número de vagas é {len(vagas['id'])}")


#with open('soup_content.txt', 'w', encoding='utf-8') as f:
#    f.write(str(soup))
```
